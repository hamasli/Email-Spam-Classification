```python
import pandas as pd;
import numpy as np;
import sklearn as sk
from sklearn.model_selection import train_test_split;
```

```python
df=pd.read_csv('emails.csv');
df
```

**After loading the dataset now we will apply preprocessing on the dataset**

```python
print(df.info());
```

```python
column_names=df.columns.tolist();
column_names
```

```python
print(len(column_names))
```

```python
df=df.drop(columns=df.columns[0]);
x=df.drop(columns=['Prediction']);
y=df['Prediction'];
print(y);
print(df)
```

```python
print(x.isnull().sum().sum());
```

```python
#here will will write 0 at all the places where column row values are null
x=x.fillna(0);
print(x.isnull().sum().sum());
```

```python
y=y.fillna(0);
print(y.isnull().sum());
```

```python
# df.shape
print(df['Prediction'].value_counts())
```

now you can see that there is no missing value in dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42);
print("Training set size:", x_train.shape[0])
print("Testing set size:", x_test.shape[0])
```

```python
from sklearn.feature_selection import SelectKBest, chi2

selector = SelectKBest(chi2, k=1000)  # Choose the top 1000 features
X_train_selected = selector.fit_transform(x_train, y_train)
X_test_selected = selector.transform(x_test)

print("Selected features shape:", X_train_selected.shape)
```

You can see that we have performed the feature selection on our dataset and we have selected only top 1000 features

**Now we will build models and train them on training dataset.**

```python
# Initialize Naive Bayes Multinomial model
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

nb_multinomial = MultinomialNB()
nb_multinomial.fit(X_train_selected, y_train)
```

```python

# Initialize Naive Bayes Gaussian model
from sklearn.naive_bayes import GaussianNB


nb_gaussian = GaussianNB()
nb_gaussian.fit(X_train_selected, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize Decision Tree (J48) model
j48 = DecisionTreeClassifier()
j48.fit(X_train_selected, y_train)
```

```python
# Predict using Naive Bayes Multinomial model
y_pred_nb_multinomial = nb_multinomial.predict(X_test_selected)

print("Naive Bayes Multinomial Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_nb_multinomial))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_nb_multinomial))
print("Classification Report:\n", classification_report(y_test, y_pred_nb_multinomial))
```

```python
# Predict using Naive Bayes Gaussian model
y_pred_nb_gaussian = nb_gaussian.predict(X_test_selected)

# Evaluate the model
print("Naive Bayes Gaussian Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_nb_gaussian))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_nb_gaussian))
print("Classification Report:\n", classification_report(y_test, y_pred_nb_gaussian))
```

```python
# Predict using Decision Tree (J48) model
y_pred_j48 = j48.predict(X_test_selected)

# Evaluate the model
print("Decision Tree (J48) Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_j48))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_j48))
print("Classification Report:\n", classification_report(y_test, y_pred_j48))
```

*You can see the performace of models on unseen(training data). We got a really god accuracy on all models greater than 0.90. And the best fit model on our dataset is Naive Bayes Multinomial model with the accuracy of 0.94*
